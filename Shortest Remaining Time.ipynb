```python
import random
```

```python
array=[[],[],[],[],[]]

for i in range(0,5):
    for a in range(random.randint(4,10)):
        b=random.randint(1,8)
        array[i].append(b)
        #print(b)
print("Five Random Arrays of length between 4 to 10 : \n" ,array)
```

```python

while len(array)>0:
    
    for x in array[0][:]:
        #array.sort(key=len)
        print("Sorted array: \n",array)
        del array[0][0]
        print("\nDeleted Item: ",x)
        array = list(filter(None, array)) 
        array.sort(key=len)
        print(array,'\n')
    
print("array is empty")
```
